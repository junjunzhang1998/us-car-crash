```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("../data/processed/crss_panel_2019_2023.csv")
df.shape
```

```python
for c in ["INJSEV_IM", "INJ_SEV"]:
    if c in df.columns:
        print(c)
        print(df[c].value_counts(dropna=False).sort_index())
```

We look at the INJSEV_IM column, where the injury severity follows this ordered logic:

- 0  → No injury
- 1  → Possible injury
- 2  → Non-incapacitating injury
- 3  → Incapacitating injury
- 4  → Fatal injury
- 5  → Injured–Severity Unknown
- 6  → Died Prior to Crash

```python
#Dropping 5 and 6 in severity levels- not useful and not large in numbers anyways
df = df[df["INJSEV_IM"].isin([0,1,2,3,4])]
```

```python
#Creating target model- severe injuries in car crushes level 3,4
df["TARGET_SEVERE"] = df["INJSEV_IM"].isin([3,4]).astype(int)
df["TARGET_SEVERE"].value_counts(normalize=True)
```

Based on 2019- 2023 data, 
- ~6.0% of drivers in accidents had severe injuries; 
- ~94.0% non-severe

```python
#export data to csv
df.to_csv(
    "../data/processed/crss_modeling_severe.csv",
    index=False
)
```

```python
import pandas as pd
from pathlib import Path

BASE = Path("../data/raw")
years = [2019, 2020, 2021, 2022, 2023]

def cols_for(year, fname):
    fp = BASE / str(year) / fname
    df = pd.read_csv(fp, nrows=5)  # just peek
    return set(df.columns)

person_cols = {y: cols_for(y, "PERSON.csv") for y in years}
vehicle_cols = {y: cols_for(y, "VEHICLE.csv") for y in years}
acc_cols = {y: cols_for(y, "ACCIDENT.csv") for y in years}

# columns common to all years
common_person = set.intersection(*person_cols.values())
common_vehicle = set.intersection(*vehicle_cols.values())
common_acc = set.intersection(*acc_cols.values())

len(common_person), len(common_vehicle), len(common_acc)
```
